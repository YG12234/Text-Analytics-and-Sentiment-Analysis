**Remove Stop Words and Punctuation**

```python
import nltk
import string
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords

# Pre-define some variables
TEXT = """Textual information in the world can be broadly categorized into two main types:
facts and opinions. Facts are objective expressions about entities, events, and their properties.
Opinions are usually subjective expressions that describe people’s sentiments, appraisals, or feelings
toward entities, events, and their properties."""

# Tokenize the text into words using the NLTK word_tokenize() function
words = nltk.word_tokenize(TEXT)

# Remove the punctuation from the words
words = [word for word in words if word not in string.punctuation]

# Remove stop words and display them
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]
output = ' '.join(filtered_words)
print("Stop words found: ", stop_words.intersection(set(words)))
print("Filtered words: ", output)
```

**Q2. Form word stemming:**


**Regular Expression**

```python
import re
from nltk.stem import PorterStemmer, LancasterStemmer

text = """Textual information in the world can be broadly categorized into two main types:
facts and opinions. Facts are objective expressions about entities, events, and their properties.
Opinions are usually subjective expressions that describe people’s sentiments, appraisals, or feelings
toward entities, events, and their properties."""

# Regular Expression
stemmer_re = re.compile(r'(ing|ly|ed|ious|ies|ive|es|s|ment)?$')
tokens_re = text.split()
stemmed_re = [stemmer_re.sub('', token) for token in tokens_re]

print("\nStemming using Regular Expression:")
print(stemmed_re)
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import RegexpTagger

text = "The big black dog barked at the white cat and chased it away."
words = word_tokenize(text)
patterns = [
    (r'\b[A-Z][a-z]*\b', 'NOUN'),
    (r'\b\d+\b', 'CARDINAL NUMBERS'),
    (r'\b\w+ing\b', 'GERUNDS'),
    (r'\b\w+ed\b', 'PAST TENSE'),
    (r'\b\w+ly\b', 'ADVERB'),
    (r'\b[\w-]+\b', 'NOUN'),
    (r'\b[A-Za-z]+\b', 'NOUNN'),
    (r'\b[^\d\W]+\b', 'ABJECTIVE'),
    (r'.*', 'NOUN')
]
regexp_tagger = RegexpTagger(patterns)
tagged_words = regexp_tagger.tag(words)
print(tagged_words)
```

**Q4. Work on sentence probabilities**

Unsmooth

```python
from collections import Counter

# Training corpus
corpus = [
    "<s> He read a book </s>",
    "<s> I read a different book </s>",
    "<s> He read a book by Danielle </s>"
]

# Tokenize the corpus
tokenized_corpus = [sentence.split() for sentence in corpus]

# Compute bigram counts
bigrams = [(tokenized_corpus[i][j], tokenized_corpus[i][j+1]) for i in range(len(tokenized_corpus)) for j in range(len(tokenized_corpus[i])-1)]
bigram_counts = Counter(bigrams)

# Unsmoothed bigram language model
def unsmoothed_bigram_probability(sentence):
    tokens = sentence.split()
    probability = 1.0
    for i in range(len(tokens)-1):
        count_bigram = bigram_counts.get((tokens[i], tokens[i+1]), 0)
        count_unigram = tokenized_corpus.count(tokens[i])
        if count_unigram == 0:
            probability = 0.0
            break
        probability *= count_bigram / count_unigram
    return probability

# Sentence to evaluate
test_sentence = "<s> I read a book by Danielle </s>"

# Calculate unsmoothed probability
unsmoothed_prob = unsmoothed_bigram_probability(test_sentence)

print("Unsmoothed bigram probability:", unsmoothed_prob)
```

Smooth

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.lm import Laplace
from nltk.lm.preprocessing import padded_everygram_pipeline

# Training corpus
corpus = [
    'He read a book',
    'I read a different book',
    'He read a book by Danielle'
]

# Tokenize the corpus
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]

# Preprocessing and generating bigrams
train_data, vocab = padded_everygram_pipeline(2, tokenized_corpus)

# Train the Laplace smoothed bigram language model
lm = Laplace(2)
lm.fit(train_data, vocab)

# Sentence to calculate probability for
sentence = ['<s>', 'I', 'read', 'a', 'book', 'by', 'Danielle', '</s>']

# Calculate probability
probability = lm.score(sentence)
print("Sentence Probability:", probability)
```

**Text Preprocessing**

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.stem import SnowballStemmer

# Initialize the SnowballStemmer
stemmer = SnowballStemmer('english')

# Define the text preprocessing function with stemming
def preprocess_text(text):
    # Tokenize the text
    words = nltk.word_tokenize(text.lower())
    # Remove stop words
    words = [word for word in words if word not in nltk.corpus.stopwords.words('english')]
    # Stem the words
    words = [stemmer.stem(word) for word in words]
    # Join the words back into a single string
    return ' '.join(words)

# Add the preprocess_text function to the dataframe
hotel_review["Review"] = hotel_review["Review"].apply(preprocess_text)
```

**Feature Extraction Using TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# Fit and transform the training data
hotel_review_TFIDF = vectorizer.fit_transform(hotel_review["Review"])
hotel_review_TFIDF
```

**Oversampling Using SMOTE**

```python
from sklearn.utils import resample
from collections import Counter

# Use SMOTE to oversample
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(hotel_review_TFIDF,hotel_review["Sentiment"])

print("Before Upsampling:-")
print(Counter(hotel_review["Sentiment"]))

print("After Upsampling:-")
print(Counter(y))
```

**Split Data**

```python
import sklearn.model_selection as model_selection
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Model Training**

**Multinomial Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

nb = MultinomialNB()

nb.fit(X_train, y_train)
y_pred_train = nb.predict(X_train)
y_pred_test = nb.predict(X_test)

print("Training Accuracy score: "+str(round(accuracy_score(y_train,y_pred_train),4)))
print("Testing Accuracy score: "+str(round(accuracy_score(y_test,y_pred_test),4)))
print(classification_report(y_test, y_pred_test, target_names=['0', '1', '2']))
```

**Hyperparameter Selection for Multinomial Naive Bayes Using GridSearch**

```python
from sklearn.model_selection import GridSearchCV


# Define the parameter grid for GridSearchCV
param_grid = {
    'alpha': [0.01, 0.1, 0.5, 1.0, 10.0, 100.0],
    'fit_prior': [True, False],
}

# Initialize GridSearchCV
nb_grid = GridSearchCV(MultinomialNB(), param_grid, scoring='accuracy', n_jobs=-1)

# Fit GridSearchCV to the training data
nb_grid.fit(X_train, y_train)

# Print the best parameters
print("Best Parameters: ", nb_grid.best_params_)

# Predict the labels for the test set using the best model parameters
y_pred_train_grid = nb_grid.predict(X_train)
y_pred_test_grid = nb_grid.predict(X_test)

# Print the accuracy scores using the best parameters
print("Training Accuracy Score: "+str(round(accuracy_score(y_train,y_pred_train_grid),4)))
print("Testing Accuracy Score: "+str(round(accuracy_score(y_test,y_pred_test_grid),4)))
print(classification_report(y_test, y_pred_test_grid, target_names=['0', '1', '2']))
```
